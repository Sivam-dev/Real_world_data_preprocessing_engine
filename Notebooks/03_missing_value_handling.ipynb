Final Missing Value Handling Strategy

After analyzing the distribution, skewness, and missing percentage of each feature, the following imputation strategies have been finalized:

🔹 Numerical Features

Skewed distributions → Median imputation

Approximately symmetric distributions → Mean imputation

Zero-inflated features (majority values = 0) → Median imputation

Justification:
Median is robust to outliers and skewed data, while mean is appropriate only when the distribution is reasonably symmetric and not affected by extreme values.

🔹 Categorical Features

Low to moderate missing percentage → Most Frequent (Mode) imputation

Features where missing may represent an implicit category → Constant value imputation ("Unknown")

Justification:
Mode imputation preserves the dominant category distribution.
Constant imputation prevents distortion when missingness may carry structural meaning.

🔹 Implementation Plan

These strategies will be implemented in the final preprocessing engine using:

SimpleImputer(strategy="mean")

SimpleImputer(strategy="median")

SimpleImputer(strategy="most_frequent")

SimpleImputer(strategy="constant", fill_value="Unknown")

All imputations will be applied through a structured Pipeline and ColumnTransformer to ensure consistency, automation, and reusability.
