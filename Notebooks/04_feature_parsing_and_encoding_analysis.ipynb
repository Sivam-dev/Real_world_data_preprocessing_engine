```python
import pandas as pd

columns = [
    "age", "workclass", "fnlwgt", "education", "education_num",
    "marital_status", "occupation", "relationship", "race", "sex",
    "capital_gain", "capital_loss", "hours_per_week", "native_country",
    "income"
]

df = pd.read_csv(
    "adult_income_raw.csv",
    sep=", ",
    engine="python",
    header=None,
    names=columns,
    na_values="?"
)
print("the dataset is loaded successfully ")
```

                                                    FEATURE SEMANTIC INTERPRETETION 
                                                ---------------------------------------

COLUMN_NAME        | PANDAS_DTYPE | SEMANTIC_TYPE | ISSUES
-------------------|--------------|---------------|----------------------------------------------
age                | object       | Numerical     | Parsing issue: stored as object instead of numeric.
workclass          | object       | Categorical   | Parsed correctly, but high missingness may bias encoding.
fnlwgt             | float64      | Numerical     | Census weight; may distort learning if treated as a regular feature.
education          | object       | Categorical   | Parsed correctly; overlaps with education_num.
education_num      | float64      | Numerical     | Encoded representation of education; redundancy risk with education.
marital_status     | object       | Categorical   | Parsed correctly; no direct issue.
occupation         | object       | Categorical   | Heavy missingness; encoding may amplify noise.
relationship       | object       | Categorical   | Parsed correctly; no direct issue.
race               | object       | Categorical   | Sensitive attribute; ethical and bias considerations required.
sex                | object       | Categorical   | Sensitive attribute; may introduce bias if not analyzed carefully.
capital_gain       | float64      | Numerical     | Highly skewed distribution; scale dominance risk.
capital_loss       | float64      | Numerical     | Highly skewed distribution; scale dominance risk.
hours_per_week     | float64      | Numerical     | Parsed correctly; no direct issue.
native_country     | object       | Categorical   | High missingness and high cardinality.
income             | object       | Categorical   | Target variable; not used as input feature.

  ---->Now what about the encoding strategies and the transformation strategies that should be applied in this dataset ,when the column is ordinal then ordinal encoding will be applied and if the column is nominal OneHotEncoder will be applied

                                                Encoding & Transformation Strategy
                                            -------------------------------------------
COLUMN_NAME        | CATEGORY_TYPE | ENCODING / TRANSFORMATION | REASON
-------------------|---------------|----------------------------|--------------------------------------------
age                | Numerical     | Convert to numeric + scale | Age is a number but stored as text.
workclass          | Nominal       | One-Hot Encoding           | Categories have no order.
fnlwgt             | Numerical     | Drop                       | Not a real feature; used only for survey weighting.
education          | Nominal       | One-Hot Encoding           | Education names have no order.
education_num      | Ordinal       | Keep as numeric            | Represents increasing education levels.
marital_status     | Nominal       | One-Hot Encoding           | Categories do not have ranking.
occupation         | Nominal       | One-Hot Encoding           | Job types have no order.
relationship       | Nominal       | One-Hot Encoding           | Relationship types have no order.
race               | Nominal       | One-Hot Encoding           | Categories are labels, not ranked.
sex                | Nominal       | Binary Encoding            | Only two categories.
capital_gain       | Numerical     | Scale                      | Numeric values with large range.
capital_loss       | Numerical     | Scale                      | Numeric values with large range.
hours_per_week     | Numerical     | Scale                      | Numeric feature on different scale.
native_country     | Nominal       | One-Hot Encoding           | Country names have no order.
income             | Nominal       | Label Encoding (target)    | Target variable for prediction.

                                                Missing Value Handling Strategy 
                                            ---------------------------------------
->Columns with no missing values will not be modified in this step.

->Columns with trivial missing values will be handled using simple imputation methods in later notebooks.

->Columns with heavy missing values will be handled carefully by either:
  treating missing values as a separate category, or
  applying frequency-based imputation, depending on model behavior.

->Rows with missing values in the target column will be removed.
